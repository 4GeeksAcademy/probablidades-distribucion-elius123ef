# Ejercicios de Random Variables

```python
from scipy.stats import binom, norm, uniform
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

## Binomial distribution

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
def dbinom(x, size, prob=0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k=x, n=size, p=prob, loc=0)
    return result


def pbinom(q, size, prob=0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k=q, n=size, p=prob, loc=0)
    return result


def qbinom(p, size, prob=0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q=p, n=size, p=prob, loc=0)
    return result


def rbinom(n, size, prob=0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n=size, p=prob, size=n)
    return result
```

```python
data = rbinom(2000, size=10, prob=0.2)
pmf = dbinom(data, size=10, prob=0.2)
cdf = pbinom(data, size=10, prob=0.2)
ppf = qbinom(data, size=10, prob=0.2)
```

```python
fig, axis = plt.subplots(3, 1, figsize=(5, 7))
axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

plt.tight_layout()  # Adjust the layout
plt.show()  # Show the plot
```

### Distribucion normal

```python
def dnorm(x, mean=0, std=1):
    """
    Calcula la distribucion normal
    """
    result = norm.pdf(x, loc=mean, scale=std)

    return result

def pnorm(x, mean=0, std=1):
    """
   Calcula la acumulativa de la distribucion normal
    """
    result = norm.cdf(x = x, loc = mean, scale = std)

    return result

def qnorm(p, mean =0, std=1):
    """
    Calcula la cuantil de la distribucion normal
    """
    result = norm.ppf(q=p, loc=mean, scale=std)

    return result

def rnorm(n, mean=0, std=1):
    """
    Genera variables random de la distribucion normal
    """
    result = norm.rvs(size=n, loc=mean, scale=std)

    return result
```

### Visualizamos en grafico

```python
mean = 0
std = 1
data = np.arange(-5, 5, 0.15)

pmf = dnorm(data, mean=mean, std =std)
cdf = pnorm(data, mean=mean, std=std)
ppf = qnorm(data, mean=mean, std=std)

fig, axis = plt.subplots(3, 1, figsize=(5, 8))

axis[0].stem(data, pmf, linefmt="r-", markerfmt="ro", basefmt=" ")
axis[1].stem(data, cdf, linefmt="b-", markerfmt="bo", basefmt=" ")
axis[2].stem(data, ppf, linefmt="g-", markerfmt="go", basefmt=" ")

plt.show()
```

### Distribucion uniforme

```python
def dunif(x, low=0, high=1):
    """
    Calculamos el punto estimado de la distribucion uniforme
    """
    result = uniform.pdf(x, loc=low, scale=(high - low))

    return result

def punif(q,low=0, high=1):
    """
    Calculamos la acumulativa de la distribucion uniforme
    """
    result = uniform.cdf(q, loc=low, scale=(high - low))

    return result

def qunif(p, low=0, high=1):
    """
    Calculamos la cuantil de la distribucion uniforme
    """
    result = uniform.ppf(p, loc=low, scale=(high - low))

    return result

def runif(n, low=0, high=1):
    """
    Generamos variables random de la distribucion uniforme
    """
    result = uniform.rvs(loc=low, scale=(high - low), size=n)

    return result
```

### Visualizamos en grafico

```python
low = 0
high = 1
data = np.arange(-5, 5, 0.15)

pmf = dunif(data, low= low, high=high)
cdf = punif(data, low=low, high=high)
ppf = qunif(data, low=low, high=high)

fig, axis = plt.subplots(3, 1, figsize=(7, 8))

axis[0].stem(data, pmf, linefmt="r-", markerfmt="ro", basefmt=" ")
axis[1].stem(data, cdf, linefmt="b-", markerfmt="bo", basefmt=" ")
axis[2].stem(data, ppf, linefmt="g-", markerfmt="go", basefmt=" ")

plt.show()
```
